```python
import xml.etree.ElementTree as ET

# Load XML file
tree = ET.parse("sample2.xml")  
root = tree.getroot()  # Get the root element

# Print the root tag
print(root.tag)
```

```python
for child in root:
    print(child.tag, child.attrib)  # Print tag name and attributes
```

```python
for elem in root.findall("element_name"):  # Replace with actual XML tag
    print(elem.text)  # Print the text inside the tag
```
